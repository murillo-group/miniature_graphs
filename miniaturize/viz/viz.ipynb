```python
# Import modules
import matplotlib.pyplot as plt
import viz
import numpy as np
import functools
```

```python
# Load rc parameters from a file
viz.initialize("sheets/viz.mplstyle")


# Generate some sample data
x = np.linspace(0, 10, 100)
y1_line = np.sin(x)
y2_line = np.cos(x)
y3_line = np.sin(x) * np.cos(x)

x_scatter = np.random.rand(100)
y3_scatter = np.random.rand(100)
y4_scatter = np.random.rand(100)
y5_scatter = np.random.rand(100)

data1_hist = np.random.normal(0, 1, 1000)
data2_hist = np.random.normal(1, 1.5, 1000)
data3_hist = np.random.normal(2, 0.5, 1000)

# Create a figure and subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Line Plot
axs[0].plot(x, y1_line, label='sin(x)')
axs[0].plot(x, y2_line, label='cos(x)')
axs[0].plot(x, y3_line, label='sin(x)*cos(x)')

# Scatter Plot
axs[1].scatter(x_scatter, y3_scatter, label='Group 1')
axs[1].scatter(x_scatter, y4_scatter, label='Group 2')
axs[1].scatter(x_scatter, y5_scatter, label='Group 3')
axs[1].legend()

# Histogram
axs[2].hist(data1_hist, bins=30, alpha=0.5, label='Data 1')
axs[2].hist(data2_hist, bins=30, alpha=0.5, label='Data 2')
axs[2].hist(data3_hist, bins=30, alpha=0.5, label='Data 3')
axs[2].legend()

# Adjust layout
plt.tight_layout()

# Save the plots to a PDF file
plt.savefig('plots.pdf')

# Show the plots
plt.show()

viz.finalize()
```
